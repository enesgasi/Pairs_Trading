```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('sample_data.csv')
```

```python
df
```

```python
plt.rcParams["figure.figsize"] = [20, 5]
plt.rcParams["figure.autolayout"] = True

price1 = df['price1'].head(100)
price2 = df['price2'].head(100)

plt.plot(price1, color='r', label='Price 1') 
plt.plot(price2, color='g', label='Price 2') 

plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [20, 5]
plt.rcParams["figure.autolayout"] = True
profit = df['profit']

plt.plot(profit)
```
